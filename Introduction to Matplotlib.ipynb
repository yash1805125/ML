# Matplotlib example workflow

```python
#import matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt

#1. get data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot

fig, ax = plt.subplots(figsize=(10,10))  #width and height of plot

#3. Plot Data
ax.plot(x,y)

#4. Customize plots
ax.set(title="Simple plot",xlabel="x-axis",ylabel="y-axis")

#5. Save and  show(you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#Make a plot from dictionary

nut_butter_prices = {"Almond butter":10,
                    "Peanut butter":8,
                     "Cashew butter":12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
     ylabel="Prices($)");
```

```python
fig ,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
#Histogram
#Make data
x = np.random.randn(1000)
```

```python
x
```

```python
fig ,ax=plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
#subplot 1
fig ,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(10));
```

```python
#Subplot option 2

fig ,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax[1,1].hist(y);
```
